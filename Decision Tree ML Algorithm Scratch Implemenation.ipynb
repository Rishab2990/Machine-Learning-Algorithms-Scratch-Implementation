```python
dataset=[
    [25,  'Aayush',      'Bought_Laptop Yes'],
    [26,  'Raghav',      'Not_Bought No' ],
    [24,  'Pranav',      'Bought_Laptop Yes'],
    [28,  'Prasant',     'Not_Bought No'],
]

header = ["Age", "Student", "Buyer"]


# Count how many people  bought or didn't buy 1st part
bought_count=0 
not_bought_count =0
buyers_info=[]

for column in dataset:
     buyers_info = column[2]
    
     if  "Yes" in buyers_info:
        bought_count+=1

     elif "No" in buyers_info:
        not_bought_count+=1


        
print("Laptop Bought : " , bought_count)
print("Laptop Not Bought : " , not_bought_count)

print() # providing space 

# Entropy Calculation of Impurity in data           2nd part 
  # (i) - find the total number of people

total_people = bought_count+not_bought_count

 # (ii) - calculate the probability of yes and no

probability_yes = bought_count/total_people
probability_no  = not_bought_count/total_people

print("Total number of people : " , total_people)
print("Probability of yes : " , probability_yes)
print("Probability of no : " , probability_no)

print() # providing space 


#this is first final calculation of entropy             
 # (iii) formula of entropy -(pYes.log2(pYes)-(pNo.log2(pNo)) in this formula (pYes) and (pNo) is a probibility that we calculated log2 = 1 in this case

final_entropy = -(probability_yes*(-1))-(probability_no*(-1))
print("The impurity of data entropy : " , final_entropy)


#splitting the data into two parts    3rd part

Yes = []
No = []
print("The datatype of 'Yes' variable : ",type(Yes))

# (i). Group 1 = age <= 25
# (ii). Group 2 = age > 25

for row in dataset:
    Age = row[0]
    if Age <=25:
      Yes.append(row)

    else:
     No.append(row)
print("people with age less than 25 :" , Yes)
print("people with age greater than 25 :" , No)


#After splitting the data again, calculate how many "Yes" and "No" are Present in the split data  ####Important because everytime we will do this until get a pure data and 0 entropy or gini impurity
# This is "Yes" part dataset after split

print() # providing space 

new_yes_count1 = 0  # I used  int instead of list, otherwise I get an error: argument type 'int' is not iterable in Python. This error occurs when we try to use two different data types, like int and list, and others
new_no_count1 = 0

for split_Data_Value1 in Yes:
    
    new_Split_value1 = split_Data_Value1[2]

    if "Yes" in new_Split_value1:
        new_yes_count1+=1

    else:
        new_no_count1+=1

print("After splitting the data we get Total in Yes part, Yes value  : ",new_yes_count1 )
print("After splitting the data we get Total in Yes part, No value   : ",new_no_count1 )


      

#This is the "No" part dataset after the split        

new_yes_count2 = 0
new_no_count2  = 0
split_Data_Value2 = []

for split_Data_Value2 in No:
    new_Split_value2 = split_Data_Value2[2]

    if "Yes" in new_Split_value2:
        new_yes_count2+=1

    else:
        new_no_count2+=1

print("After splitting the data we get Total in No part, Yes value  : ",new_yes_count2 )
print("After splitting the data we get Total in No part, No value   : ",new_no_count2 )


# Now again, calculate the entropy of each group, Yes and No
# '''(i) Calculating entropy for Yes group . i have question like in this code i've done split with the data, but there is something is which is may be a
# problem like there is now there is a value which have less than age 25 and greater than age 25, but inside that specific group, only two entities or 
# data value are present that means the etropy of both yes and not group were 0 because according to the formula, there is yes group and there is no group 
# , in the no group we have 0 value as an input '''

#Again repeat same step like previous one , calculating probability before calculating etropy

print() # providing space 

# Yes part probability less than 25 age
total_count1 = new_yes_count1 + new_no_count1

bought_probability = new_yes_count1/total_count1
no_probability = new_no_count1/total_count1
print("Yes part bought probability : ", bought_probability)
print("Yes part not bought probability : ", no_probability)


# No part probability greater than 25 age
total_count2 = new_yes_count2 + new_no_count2

bought_probability1 = new_yes_count2/total_count2
no_probability1 = new_no_count2/total_count2

print("Yes part bought probability : ", bought_probability1)
print("Yes part not bought probability : ", no_probability1)

print() # providing space 

# Now we will calculate Entropy using splitted data probility numbers 
yes_splitted_entropy = -(0*(-0)+0*(-0))
print("yes splitted data entropy calculation :", yes_splitted_entropy)

print()

no_splitted_entropy1 = -(0*(-0)+ 0*(-0))
print("no splitted data entropy calculation :", no_splitted_entropy1)

```
